```python
import pandas as pd
import os

# Define the file path
file_path = '../Desktop/Startup-Success-Forecasting-Framework/data/Merged_Successful_V2'

# Check if the file exists
if not os.path.exists(file_path):
    print(f"Error: The file {file_path} does not exist.")
else:
    try:
        # Read the CSV file
        df = pd.read_csv(file_path)
        
        # Print the first few rows
        print("First few rows of the dataset:")
        print(df.head())
        
        # Print some basic information about the dataset
        print("\nDataset Info:")
        print(f"Number of rows: {df.shape[0]}")
        print(f"Number of columns: {df.shape[1]}")
        print("\nColumn names:")
        print(df.columns.tolist())
        
        # Print short_description and long_description for each row
        print("\nShort and Long Descriptions:")
        for index, row in df.iterrows():
            print(f"\nRow {index + 1}:")
            print(f"Short Description: {row['short_description']}")
            print(f"Long Description: {row['long_description']}")
            print("-" * 50)  # Print a separator line
            
            # Optional: limit the number of rows printed
            if index == 4:  # This will print descriptions for the first 5 rows
                break
        
    except pd.errors.EmptyDataError:
        print("The file is empty.")
    except pd.errors.ParserError:
        print("Error parsing the CSV file. Please check if it's a valid CSV.")
    except Exception as e:
        print(f"An error occurred while reading the file: {str(e)}")
```
